```python
import numpy as np
import pandas as pd
import keras
import cv2
from matplotlib import pyplot as plt
import os
import random
from PIL import Image
from keras.applications.vgg19 import VGG19
from keras.layers import Dense, BatchNormalization
from keras.models import Sequential
from keras.optimizers import RMSprop
from sklearn.preprocessing import LabelEncoder

# Load the dataset CSV
df = pd.read_csv("./dataset/train.csv")
df.head()  # Display the first few rows of the dataset

# Check the column names to confirm 'landmark_id' exists
print(df.columns)

# Use the correct column name based on the output
# If 'landmark_id' exists, continue as planned; if not, adjust the name below accordingly

# Limit dataset for faster exploration
samples = 50
df = df.loc[:samples, :]

# Count unique classes (landmark IDs)
if 'landmark_id' in df.columns:
    num_classes = len(df["landmark_id"].unique())
    num_data = len(df)

    # Display distribution of landmark IDs
    data = pd.DataFrame(df["landmark_id"].value_counts()).reset_index()
    data.columns = ['index', 'landmark_id']
else:
    print("'landmark_id' column not found in the DataFrame.")
# Plot histogram of landmark IDs
plt.hist(data['landmark_id'], bins=50, range=(0, 58), label='Landmarks')
plt.title('Landmark ID Distribution')
plt.show()

# Encode landmark IDs
lencoder = LabelEncoder()
lencoder.fit(df["landmark_id"])

# Functions for encoding and decoding labels
def encode_label(lb1):
    return lencoder.transform(lb1)

def decode_label(lb1):
    return lencoder.inverse_transform(lb1)

# Load image based on index number
def get_image_from_number(num):
    fname = df.loc[num, "url"]  # Use the local file path directly from the CSV
    label = df.loc[num, "landmark_id"]
    
    # Read image using the provided path
    im = cv2.imread(fname)
    if im is None:
        raise FileNotFoundError(f"Image not found: {fname}")
    
    return im, label

# Display sample images from random samples
fig = plt.figure(figsize=(16, 16))

try:
    for i in range(1, 5):
        # Select a random index
        rand_index = random.randint(0, len(df) - 1)
        
        # Load and display the image
        img, _ = get_image_from_number(rand_index)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display
        fig.add_subplot(1, 4, i)
        plt.imshow(img)
        plt.axis('off')

except ValueError as e:
    print(e)

plt.show()
# Parameters for training
learning_rate = 0.0001
loss_function = "sparse_categorical_crossentropy"

# Load VGG19 without the top classification layer
source_model = VGG19(weights=None, include_top=False, input_shape=(224, 224, 3))

# Build the new model
model = Sequential()
for layer in source_model.layers:
    model.add(layer)

model.add(BatchNormalization())
model.add(Dense(num_classes, activation="softmax"))

model.summary()

# Compile the model
optimizer = RMSprop(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss=loss_function, metrics=['accuracy'])

# Function to resize images to the target size
def image_reshape(img, target_size):
    return cv2.resize(img, target_size)

# Get batch of data
def get_batch(dataframe, start, batch_size):
    image_array = []
    label_array = []

    end_img = start + batch_size
    if end_img > len(dataframe):
        end_img = len(dataframe)

    for idx in range(start, end_img):
        im, label = get_image_from_number(idx)
        im = image_reshape(im, (224, 224))
        image_array.append(im)
        label_array.append(label)

    label_array = encode_label(label_array)

    return np.array(image_array), np.array(label_array)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, BatchNormalization
from tensorflow.keras.applications import VGG19
from tensorflow.keras.optimizers import RMSprop

# Define image dimensions
image_height = 224
image_width = 224
channels = 3

# Load the VGG19 model without the top layer
source_model = VGG19(weights=None, include_top=False, input_shape=(image_height, image_width, channels))

model = Sequential()
for layer in source_model.layers:
    model.add(layer)

# Add layers to flatten the output and classify
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(num_classes, activation="softmax"))

# Compile the model
optimizer = RMSprop(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Training loop
epochs = 5
batch_size = 16

for e in range(epochs):
    print(f"Epoch: {e+1}/{epochs}")
    train = df.sample(frac=1).reset_index(drop=True)  # Shuffle the data
    
    for it in range(int(np.ceil(len(train) / batch_size))):
        X_train, y_train = get_batch(train, it * batch_size, batch_size)
        print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
        
        # Train the model on the batch
        model.train_on_batch(X_train, y_train)

# Save the trained model
model.save("landmark_detection_model.keras")


import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model

# Load the trained model
model = load_model("landmark_detection_model.keras")

# Function to load and preprocess a test image
def load_and_preprocess_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Image not found: {image_path}")
    
    img = image_reshape(img, (224, 224))  # Resize image
    img = img.astype("float32") / 255.0  # Normalize image
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img

# Function to predict the landmark
def predict_landmark(image_path):
    img = load_and_preprocess_image(image_path)
    predictions = model.predict(img)
    predicted_class = np.argmax(predictions, axis=1)  # Get the index of the max probability
    return decode_label(predicted_class)[0], img  # Decode the label back to original

# Select a random test image from the CSV
random_index = random.randint(0, len(df) - 1)
test_image_path = df.loc[random_index, 'url']  # Get the image path from the DataFrame

try:
    predicted_landmark_id, predicted_image = predict_landmark(test_image_path)
    
    # Load the original test image for display
    original_image = cv2.imread(test_image_path)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display

    # Display both the original and predicted image
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.axis('off')
    plt.title('Original Image')

    # Display predicted image (which is the same as the original image)
    plt.subplot(1, 2, 2)
    predicted_image = np.squeeze(predicted_image)  # Remove the batch dimension
    plt.imshow(predicted_image)
    plt.axis('off')
    plt.title(f'Predicted Image\nLandmark ID: {predicted_landmark_id}')

    plt.show()

except Exception as e:
    print(e)
```
